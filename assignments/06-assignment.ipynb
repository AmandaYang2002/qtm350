```python
def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, num):
        if num % i == 0:
            return False
    return True

primes = []
for i in range(1, 101):
    if is_prime(i):
        primes.append(i)

print(primes)
```

Here's the refactored code using Copilot:
```python
def is_prime(num):
    if num <= 1:
        return False
    if num == 2:
        return True
    if num % 2 == 0:
        return False
    for i in range(3, int(num**0.5) + 1, 2):
        if num % i == 0:
            return False
    return True

primes = [i for i in range(1, 101) if is_prime(i)]

print(primes)
```

```python
def factorial(n):
    if n < 0:
        return "Factorial is not defined for negative numbers."
    elif n == 0 or n == 1
        return 1
    else:
        result = 1
        for i in range(2, n + 1):
            result *= i
        return result
```

6. Use Copilot to generate documentation comments for the following `R` factorial function:

```r
// write a function to calculate the factorial of a number
// n: the number for which the factorial is calculated
// returns: the factorial of the input number
// example: calculate_factorial(5) returns 120
calculate_factorial <- function(n) {
  if (n == 0) {
    return(1)
  } else {
    return(n * calculate_factorial(n - 1))
  }
}
```
