2. Why does the statement `3 is 3.0` return `False` in Python, but `3 == 3.0` returns `True`? What concepts are involved here?

The `is` operator checks if two variables refer to the same object in memory, while the `==` operator checks if the values of the two variables are equal. In this case, `3` and `3.0` are different objects in memory, so `3 is 3.0` returns `False`. However, the values of `3` and `3.0` are equal, so `3 == 3.0` returns `True`.

5. Explain the difference between `//` and `/` operators. Provide an example where the distinction is important and describe a potential error if they are used interchangeably.

The `//` and `/` operators in Python are used for division, but they behave differently when the operands are integers. The `//` operator performs floor division, which returns the integer part of the division result. The `/` operator performs true division, which returns a floating-point result.

7. _Without running the code_, what will be the value of `x` at the end of the following Python snippet? Explain the order in which operations are performed.
   
```python
x = 5
x += x * (x - 1)
```

The value of `x` at the end of the snippet will be `25`. The order of operations is as follows: given the initial value of x=5, the expression `x * (x - 1)` evaluates to `5 * (5 - 1) = 20`. Then, the expression `x += 20` is equivalent to `x = x + 20`, which results in `x = 5 + 20 = 25`.
